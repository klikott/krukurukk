```python
data = pd.read_csv('/content/Employee.csv')
```

```python
gender_counts = data['Gender'].value_counts()

# Siapkan data untuk pie chart
genders = gender_counts.index
counts = gender_counts.values
colors = ('#DB2E5F', '#62D0DB')
explode = (0, 0)

plt.title('Gender Distribution in Employee Dataset')
plt.pie(
    counts,
    labels=genders,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    explode=explode,
    shadow=True
    )

plt.show()
```

Data di atas merupakan visualisasi dari distribusi gender karyawan. Berdasarkan visualisasi di atas, dapat dilihat bahwa terdapat pie chart yang tergolong dalam dua kelompok yaitu perempuan (Female) berwarna biru dan laki-laki (Male) berwarna merah yang dinyatakan dalam persentase(%).

Dari visualisasi data ini dapat disimpulkan bahwa jumlah karyawan perempuan lebih sedikit dibandingkan laki-laki dengan persentase 40,3% untuk perempuan dan 59,7% untuk laki-laki.
